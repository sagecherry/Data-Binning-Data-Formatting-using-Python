## Data Binning & Data Formatting using Python

```python
import pandas as pd
import numpy as np
data={
    'Product':['Laptop','Mobile','Tablet','Headphones','Smartwatch','Camera'],
    'Price':[55000,20000,3000,2000,8000,45000],
    'Units_sold':[25,60,35,80,50,15]
}
df=pd.DataFrame(data)
print(df)
```

```python
#Data Binning
bins1=[0,10000,30000,60000]
labels1=['Low','Medium','High']
df['Price_Category']=pd.cut(df['Price'],bins=bins1,labels=labels1)
print(df)
```

```python
#Binning Units Sold
bins2=[0,30,60,100]
labels2=['Less Sold','Moderately Sold','Highly Sold']
df['Units_Category']=pd.cut(df['Units_sold'],bins=bins2,labels=labels2)
print(df)
```

```python
#Data Formatting
#Convert Data Type
print(df.dtypes)
df['Units Sold']=df['Units_sold'].astype(float)
print(df.dtypes)
```

```python
#Convert Text to Uppercaste
df['Product']=df['Product'].str.upper()
print(df)
```

```python
#Rounding Values
df['Price']=df['Price'].round(2)
print(df)
```

```python
#Additional Operations
#Sort Data
sorted_df=df.sort_values(by='Price')
print(sorted_df)
```

```python
#Display Unique Values
print(df['Price_Category'].unique())
```

```python
#Dataset:Food Delivery Orders
data={'order_ID':['1','2','3','4','5','6','7','8'],
      'order_value':[250,450,120,600,300,150,500,220],
      'delivery_time':[30,45,25,60,35,20,50,28],
      'distance_km':[5,8,3,10,6,2,9,4]}

df=pd.DataFrame(data)
print(df)
```

```python
#Binning Order Value
bins=[0,200,400,700]
labels=['less orders','moderate orders','high orders']
df['order_category']=pd.cut(df['order_value'],bins=bins,labels=labels)
print(df)
```

```python
#Binning Delivery Time
bins2=[0,30,45,70]
labels2=['less time','moderate time','high time']
df['delivery_category']=pd.cut(df['delivery_time'],bins=bins2,labels=labels2)
print(df)
```

```python
#Data Formatting Examples
#convert distance to float
df['distance_km']=df['distance_km'].astype(float)
print(df)
```

```python
#Convert Order_ID to Uppercase
df['order_ID']=df['order_ID'].str.upper()
print(df)
```

```python
#check data types
print(df.dtypes)
```

```python
#Unique Categories
print(df['order_category'].unique())
df
```

```python
#count of categories
print(df['order_category'].value_counts())
df
```
